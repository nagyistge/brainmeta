# Classifying NeuroSynth Tags Based on Word Embeddings

Here we will build a word2vec model with all of the neurosynth corpus, and then train/test to predict individual neurosynth terms. I originally trained models for each of rbf, linear, poly, and sigmoid kernels, but the results were essentially equivalent, and I chose to use linear so we could use the weights/support vectors in some later analysis, if desired. I am sorting by the number of True Postitives, because unfortunately, the number of papers tagged in NeuroSynth for each of the terms is pretty sparse, so when I hold out 20% for train/test, it comes down to not many papers. Sorting by the TP rate ensures that more sizeable datasets are toward the top.

```python
import pandas
pandas.set_option('display.max_rows', 4000)
classifiers = pandas.read_csv("classifiers_neurosynth_neurosynth.tsv",sep="\t",index_col=0)
classifiers = classifiers.sort(columns=["TP","accuracy","TN"],ascending=False)
classifiers
```
