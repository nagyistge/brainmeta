```python
%matplotlib inline
import nibabel as nib
import numpy as np
from glob import glob
from nilearn.plotting import plot_roi, plot_stat_map

# Here are 46 cope1.nii.gz images for a subset of HCP subjects
images = glob("/home/vanessa/Desktop/hcp/*.nii.gz")

# Function to flag outliers
def find_outliers(image,n_std=6):
  data = nib.load(image).get_data()
  mean = data.mean()
  std = data.std()
  six_dev_up = mean + n_std * std
  six_dev_down = mean - n_std*std
  high_outliers = len(np.where(data>=six_dev_up)[0])
  low_outliers = len(np.where(data<=six_dev_down)[0])
  return high_outliers,low_outliers

# For each image, how may outliers?
print "HIGH  :  LOW"
for image in images:
    high_outliers,low_outliers = find_outliers(image)
    print "%0.0f  :  %0.0f" %(high_outliers,low_outliers)
```

```python
import matplotlib
matplotlib.rcParams["figure.max_open_warning"] = 50

# Function to flag outliers
def plot_outliers(image,n_std=6):
  mr = nib.load(image)
  data = mr.get_data()
  mean = data.mean()
  std = data.std()
  six_dev_up = mean + n_std * std
  six_dev_down = mean - n_std*std
  empty_brain = np.zeros(data.shape)
  empty_brain[data>=six_dev_up] = 1
  empty_brain[data<=six_dev_down] = 1 
  outlier_nii = nib.nifti1.Nifti1Image(empty_brain,affine=mr.get_affine(),header=mr.get_header())
  plot_roi(outlier_nii)

for image in images:
    plot_outliers(image)
```
