## Impact of Threshold on Correlation - Positive and Negative Values
Trying to be simple!

 1. Start with one thresholded image, I
 2. Calculate correlation against another image, J
 3. Apply threshold
 4. Calculate correlation again
 5. Do a gazillion times, make simple statement about how it changed!

```python
nruns=1000

# Threshold cutoff percentages 
thresholds = range(53,100)

# Here we will save the averages
correlations = pandas.DataFrame(columns=["Cutoff Lower","Cutoff Upper","Base Correlation","Thresh Correlation"])

for t in range(0,len(thresholds)):

    thresh = thresholds[t]

    # Will hold original correlation, thresholded correlation
    corrs=pandas.DataFrame(columns=["No Threshold","Threshold"])
    for run in range(nruns):
        img1 = numpy.random.randn(100)*1.0
        # Add 40% noise
        img2 = (img1*0.6) + (0.4*numpy.random.randn(100)*1.0)
        # Threshold image 2
        cutoff_top = scipy.stats.scoreatpercentile(img2,thresh)
        cutoff_bot = scipy.stats.scoreatpercentile(img2,(100-thresh))
        mask = numpy.logical_or(img2>cutoff_top,img2<cutoff_bot) 
        # Base case - no thresholding of image 2
        tno = numpy.corrcoef(img1,img2)[0,1]
        # Comparison case - thresholding applied
        tyes = numpy.corrcoef(img1[mask],img2[mask])[0,1]
        corrs.loc[run] = [tno,tyes]  
    correlations.loc[t] = [cutoff_bot,cutoff_top] + corrs.mean().tolist()

# Take absolute value, we just want to know if larger or smaller number
correlations["Base Correlation"] = correlations["Base Correlation"].abs()
correlations["Thresh Correlation"] = correlations["Thresh Correlation"].abs()
change = correlations["Thresh Correlation"] - correlations["Base Correlation"]
correlations["change"] = change
correlations
```
