```python
import re
import numpy
import pandas
import scipy.stats
from scipy.stats import spearmanr, pearsonr
from mne.stats import fdr_correction
```

## Impact of Threshold on Correlation - Positive and Negative Values

 1. Start with one thresholded image, I
 2. Calculate correlation against another image, J
 3. Apply threshold
 4. Calculate correlation again
 5. Repeat for 1000 permutations, assess change

```python
nruns=1000

# Threshold cutoff percentages 
thresholds = range(53,100)

# Here we will save the averages
correlations = pandas.DataFrame(columns=["Cutoff Lower","Cutoff Upper","Spearman Base","Spearman Test","Pearson Base","Pearson Test","Pearson WRS P-Value","Spearman WRS P-Value"])

for t in range(0,len(thresholds)):

    thresh = thresholds[t]

    # Will hold original correlation, thresholded correlation
    corrs=pandas.DataFrame(columns=["Spearman Base","Spearman Test","Pearson Base","Pearson Test"])
    for run in range(nruns):
        img1 = numpy.random.randn(100)*1.0
        # Add 40% noise
        img2 = (img1*0.6) + (0.4*numpy.random.randn(100)*1.0)
        # Threshold image 2
        cutoff_top = scipy.stats.scoreatpercentile(img2,thresh)
        cutoff_bot = scipy.stats.scoreatpercentile(img2,(100-thresh))
        mask = numpy.logical_or(img2>cutoff_top,img2<cutoff_bot) 
        # Base cases - no thresholding of image 2
        spearman_base = spearmanr(img1,img2)[0]
        pearson_base = pearsonr(img1,img2)[0]
        # Comparison case - thresholding applied
        spearman_test = spearmanr(img1[mask],img2[mask])[0]
        pearson_test = pearsonr(img1[mask],img2[mask])[0]
        corrs.loc[run] = [spearman_base,spearman_test,pearson_base,pearson_test]  
    # Use Wilcox Rank Sum test to determine if distributions different
    pearson_pval = scipy.stats.ranksums(corrs["Spearman Base"], corrs["Spearman Test"])[1]
    spearman_pval = scipy.stats.ranksums(corrs["Pearson Base"], corrs["Pearson Test"])[1]    
    correlations.loc[t] = [cutoff_bot,cutoff_top] + corrs.mean().tolist() + [pearson_pval,spearman_pval]

# Take absolute value, we just want to know if larger or smaller number
correlations[["Spearman Base","Spearman Test","Pearson Base","Pearson Test"]] = correlations[["Spearman Base","Spearman Test","Pearson Base","Pearson Test"]].abs()
# Calculate overall changes
pearson_change = correlations["Pearson Test"] - correlations["Pearson Base"]
spearman_change = correlations["Spearman Test"] - correlations["Spearman Base"]
correlations["Spearman Change"] = spearman_change
correlations["Pearson Change"] = pearson_change
# Finally, correct for multiple comparisons
reject,pearson_corrected = fdr_correction(correlations["Pearson WRS P-Value"], alpha=0.001, method='indep')
reject,spearman_corrected = fdr_correction(correlations["Pearson WRS P-Value"], alpha=0.001, method='indep')
correlations["Spearman WRS FDR"] = spearman_corrected
correlations["Pearson WRS FDR"] = pearson_corrected
correlations
```
