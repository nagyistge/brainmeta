```python
import matplotlib
import numpy as np
%matplotlib inline
import nibabel as nib
from nilearn.plotting import plot_stat_map, plot_roi
from scipy.spatial.distance import pdist
from scipy.stats import norm, t

all_copes_file = "/home/vanessa/Desktop/tfMRI_LANGUAGE_STORY.nii_tstat1.nii.gz"
all_copes = nib.load(all_copes_file)
plot_stat_map(all_copes)
print "Here is our map created with randomise for all 486 subjects, for the story contrast"
```

Now let's identify the outliers greater than 6 standard deviations of the mean.

```python
# Function to flag outliers
def find_outliers(image,n_std=6):
  data = nib.load(image).get_data()
  mean = data.mean()
  std = data.std()
  six_dev_up = mean + n_std * std
  six_dev_down = mean - n_std*std
  high_outliers = len(np.where(data>=six_dev_up)[0])
  low_outliers = len(np.where(data<=six_dev_down)[0])
  return high_outliers,low_outliers

high,low = find_outliers(all_copes_file)
print "There are %s HIGH and %s LOW outliers" %(high,low)
```

```python
import matplotlib

# Function to flag outliers
def plot_outliers(image,n_std=6):
  mr = nib.load(image)
  data = mr.get_data()
  mean = data.mean()
  std = data.std()
  six_dev_up = mean + n_std * std
  six_dev_down = mean - n_std*std
  empty_brain = np.zeros(data.shape)
  empty_brain[data>=six_dev_up] = 1
  empty_brain[data<=six_dev_down] = 1 
  outlier_nii = nib.nifti1.Nifti1Image(empty_brain,affine=mr.get_affine(),header=mr.get_header())
  plot_roi(outlier_nii)

plot_outliers(all_copes_file)
```

Let's now mimic doing the t_to_z conversion - first we will see how many p values of 1 we get (which would turn to -inf)

```python
dof = 486
data = all_copes.get_data()
p_values = t.sf(data, df = dof)

# How many 1s?
print "There are %s 1s in the p value map." %(len(p_values[p_values==1]))

empty_nii = np.zeros(data.shape)
empty_nii[p_values == 1.0] = 1
nii_obj = nib.nifti1.Nifti1Image(empty_nii,affine=all_copes.get_affine(),header=all_copes.get_header())
plot_roi(nii_obj)
```

Finally, we are going to try the 0.9999 substitution.

```python
def t_to_z(image1, dof):
  data = image1.get_data()
  p_values = t.sf(data, df = dof)
  # A value of 1 will return a -inf in the map, so we must sub with 0.999999999999
  p_values[p_values==1.0] = 0.9999999999
  z_values = norm.isf(p_values)
  Z_nii = nib.nifti1.Nifti1Image(z_values,affine=image1.get_affine(),header=image1.get_header())
  return Z_nii

Z = t_to_z(all_copes,dof)
plot_stat_map(Z,title="New Z Score Map")
empty_nii = np.zeros(Z.shape)
```

```python
print "Any nans? %s" % np.isnan(Z.get_data()).any()
print "Any infs? %s" % np.isinf(Z.get_data()).any()
print "Max: %s" % np.max(Z.get_data())
print "Min: %s" % np.min(Z.get_data())
```
