# The Chunk Indifferent Ranking Algorithm 
This notebook will show an example output from the "Chunk Indifferent Ranking Algorithm," [described here](http://www.vbmis.com/bmi/share/russ/image_comparison/Chunk_Indifferent_Ranking_Algorithm.pdf). You should read this first to understand the algorithm.  The actual data frame is of size 315,840, as each of the 470 input images is assessed at 16 thresholds for each of three strategies for missing handling data (union, intersection, brain mask), and within each of those, there are 3 gold standards (contrast, task, and task_contrast). The "group" contrast is not included in this set, but will be for the final analysis.

### Variables

- image_id corresponds with just one contrast image, GRP08_TASK07_CON35, meaning that each of the gold standard rankings are specific to that image.
- thresh corresponds to the thresholds, in the set [0.00 0.50 1.00 1.50 1.65 1.75 1.70 1.85 1.80 1.96 1.90 2.00 2.58 3.02 3.50 4.00]
- strategy refers to how missing data is handled when calculated the pearson or spearman score, for each of which we have "union," "intersection," or "brain mask"
- accuracy is the value calculated by the "Chunk Indifferent Ranking Algorithm" (I am terrible at naming things)
- group refers to the "chunk" in the list. For example: Group 1 for "contrast" would be images with the same contrast. Group 2 would be everything else. The groups are drawn out in the PDF linked above.
- standard refers to how the labeling is done, for example "contrast" would place images of the same contrast in a chunk at the top, followed by everything else. These standards are also explained in the PDF above.

### Observations
This scoring algorithm seems reasonable for the following reasons:

 - I would expect to see higher accuracy for thresholds closer to 0, most definitely for the contrast standard, because we haven't changed the images, and we do (it jumps from .95 accuracy down to .67)
 - I would also expect that accuracy for contrast will be greater than for task, because it's only marginally reasonable that images of different contrasts from the same tasks should come before other tasks. We see that as well. Interestingly, however, returning the same task (preliminarily) seems more robust to changes in threshold - we start at about .85, and don't change much from that.
 - The task_contast standard would need some kind of weighted mean calculation. For task and contrast, we probably don't care about the accuracy for group 2.
 - It looks like there is a slight bug in my implementation to produce a slightly negative accuracy for some cases, and I think I'm likely off by an index of 1 somewhere... need to debug this!
