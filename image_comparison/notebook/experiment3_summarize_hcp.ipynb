# Image Comparison for Whole Brain Statistical Maps

### The Basic Story
1. We do brain imaging to understand cognitive processes, and how the brain influences health and behavior. 
2. If we understand the link between brain, behavior, health, and well-being, we can help people.
3. We need meta-analysis to synthesize all this brain imaging. With meta-analysis we can find consensus across studies, validate new results, and discover interesting patterns revealed with large data.
5. Coordinate based meta analysis was the old way of doing it, but now we have whole brain maps. We've demonstrated using whole brain maps is better.
6. Thus we need to be able to do optimized meta-analysis with whole brain maps.

### Goals of the Researcher conducting Meta-Analysis
Generally, the goal is to find similar maps from "data in the wild" that would go into some whole brain meta-analysis. I also want to be confident that my choice of similarity metric, including / not including thresholded images, and image transformation is in line with my goals as a researcher.  For example:

- *Case 1*: I have a "query" image (let's say a component from an ICA) and I know nothing else about it.  If I search against a database of whole brain maps, I want to find maps that are spatially, and/or value-wise, similar.  Those "other" images may have contrast (or some other meta-data) labels, and I could look for enrichment of these terms in the most similar maps from my result. It could be the case that I want to first search for whole brain maps with similar activation in my smaller region (eg, "intersection" strategy) for a whole brain meta-analysis, and then find a different set of "mini maps" to use like coordinate data for a coordinate based meta-analysis. 
- *Case 2*: I have a whole brain statistical map result for which I know the contrast and task, and I want to find maps that are assessing the same cognitive process to look for consensus.  The other images (mostly) don't have labels, and I want to find the right number / set to include.
- *Case 3*: I'm in the Poldracklab and very frustrated that no one wants to add contrast labels to their uploaded maps. If I have a small subset of maps with labels, I could generate probabilities that each map comes from a particular task or contrast. If I am able to generate a reasonably sized set with high confidence of a task or process, I could then do a meta-analysis. I should have some degree of confidence about the kind of maps that are returned (not having labels) with regard to task or contrast similarity.

### Questions that we don't know the answer to
- If the maps I'm comparing my result to are tiny or thresholded, does that bias what is returned?
- For Case 2: how many images can I use from my result and still have confidence they are an equivalent task or contrast? What similarity score should I "cut" at?
- Is pearson or spearman more robust to tiny or thresholded maps? Why?
- If I do a search and I am confident in my TASK, CONTRAST, can I be confident that some percentage of my top results are from an equivalent task? contrast? group?
- What has a bigger influence on similarity scores: having the same group, task, or contrast?

#### Levels of Assessing Rankings
A "ranking" instead of a list of pearson/spearman scores would look something like this:

This is a cool strategy because we can do an "overall ranking" (eg, in the example above, numbers 1 to 430. But we can ALSO score based on:

- Are the group maps at the top from the same task AND contrast?
- Are the group maps at the top from the same task? (but maybe different contrast?)
- Are the group maps at the top from the same group (but different tasks and contrasts?)

You can imagine that each of these might be a kind of search that a researcher would do in NeuroVault.  The first question is finding the absolute most similar group maps.  The second question would suggest that the contrasts aren't very unique.  The third question would hint that there is some bias with regard to the study itself (eg, scanner, error in analyses).

*Important Note* We would still be able to do all the analyses looking at the scores themselves.  However the ranking would be based on this labels, which is a much more "real world" use-case (I think) than having this arbitrary base standard based on the scores alone!
