# Overview of Experiment
We want to know if some weighting of the voxels in a group map for some task/contrast can improve the similarity calculation. One hypothesis is that the residuals (error) calculated when deriving the group map would be a useful weight, i.e., we could place higher weight on voxels with lower error because we have higher confidence in them, and this could serve to better find similar contrasts during our search.

Toward this goal, I will first investigate if there is a meaningful relationship between residuals in cope images and the cope map values.  If so, I can predict residuals from the randomise cope group map, and then possibly use the residuals as a weighting for the map. The other option is to edit the FSL randomise source code to output actual residuals. I can then test different weightings of the voxels (the residual weighting being one of them, potentially) toward returning some gold standard ordering of images (eg, same contrast first, same task first, etc.).

Although FSLs randomise does not output residuals, I was able to find a res4D.nii.gz file in the single subject directories, which minimally I can investigate to find some realtionship between the cope values and residuals. Here I will investigate this and see if I can predict the residuals from the code values.

## Looking at Residuals Left and Right
Can we use a mean?

```python
import nibabel
import matplotlib.pylab as plt
import numpy as np
%matplotlib inline
from nilearn.plotting import plot_stat_map

residuals = nibabel.load("/home/vanessa/Documents/Work/BRAINMETA/IMAGE_COMPARISON/experiment4/residuals/residuals.nii.gz" )
cope = nibabel.load("/home/vanessa/Documents/Work/BRAINMETA/IMAGE_COMPARISON/experiment4/residuals/cope.nii.gz") 
res1 = residuals.get_data()[:,:,:,0]
res2 = residuals.get_data()[:,:,:,1]
x = range(0,np.prod(res1.shape))
plt.plot(x,res1.flatten(), 'r--',x,res2.flatten(),'b--')
plt.show()
```

# Visualizing Residuals in the Brain
Where are the highest values? The lowest error values? Let's look at the residual map in the brain.

```python
# We need to restart the kernel - using the plots above is making some bug
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pylab as plt
from nilearn.plotting import plot_stat_map, plot_glass_brain, plot_img
import nibabel

residuals = nibabel.load("/home/vanessa/Documents/Work/BRAINMETA/IMAGE_COMPARISON/experiment4/residuals/residuals.nii.gz" )
cope = nibabel.load("/home/vanessa/Documents/Work/BRAINMETA/IMAGE_COMPARISON/experiment4/residuals/cope.nii.gz") 
res1 = residuals.get_data()[:,:,:,0]
res2 = residuals.get_data()[:,:,:,1]
res_avg = (res1+res2)/2

# The plot looks terrible with such a weird range of values, let's try normalizing
brain_voxels = res_avg[res_avg!=0]
res_norm = (brain_voxels - np.mean(brain_voxels)) / np.std(brain_voxels)
empty_nii = np.zeros(res_avg.shape)
empty_nii[res_avg!=0] = res_norm

res_nii = nibabel.nifti1.Nifti1Image(empty_nii,affine=cope.get_affine(),header=cope.get_header())
plot_stat_map(res_nii,title="Mean Residuals",display_mode="z",cut_coords=10)
plot_stat_map(cope,title="Cope Map",display_mode="z",cut_coords=10)
```

that wasn't so useful. Let's see if we can predict the residuals from the cope map, because then we could predict residuals from our map produced with randomise.

# Predicting Residuals from Cope


```python
from sklearn import linear_model
import random

# We need to reformat data to have one element per row
cope_dat = data.flatten()[:, np.newaxis]
res_dat = res_avg.flatten()[:,np.newaxis]

# Hold out quarter for testing
idx_train = random.sample(x,int(np.floor(0.75*len(x))))
idx_test = np.setdiff1d(x,idx_train)

X_train = cope_dat[idx_train]
X_test = cope_dat[idx_test]

Y_train = res_dat[idx_train]
Y_test = res_dat[idx_test]

# Build the model with mean residuals
lr = linear_model.LinearRegression()
lr.fit(X_train, Y_train)

print('Coefficients: \n', lr.coef_)
print("Residual sum of squares: %.2f" % np.mean((lr.predict(X_test) - Y_test) ** 2))

# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % lr.score(X_test, Y_test))

# Plot outputs
plt.scatter(X_test, Y_test,  color='black')
plt.plot(X_test, lr.predict(X_test), color='blue',linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```
