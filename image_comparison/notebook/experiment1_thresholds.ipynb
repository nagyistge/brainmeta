This is a quick notebook to visualize (check) the range of thresholds that I have defined for my experiment 1, where I will be calculating different similarity metrics to compare different transformations of unthresholded maps to their original selves. The thresholding is being done like this, with the following values:

```python
def threshold_abs(image1,thresholds=[0.0,0.5,1.0,1.5,1.65,1.7,1.75,1.8,1.85,1.9,1.96,2,2.58,3,3.5,4.0]):
  '''threshold the image at a range of Z score thresholds, including high positive and negative.'''
  thresholded = dict()
  data = image1.get_data()
  for thresh in thresholds:
    tmp = np.zeros(image1.shape)  
    tmp[np.abs(data) >= thresh] = data[np.abs(data) >= thresh]  
    new_image = nib.Nifti1Image(tmp,affine = image1.get_affine(),header=image1.get_header())
    thresholded[thresh] = new_image
  return thresholded
```

I am using an absolute value to get positive and negative values, and I chose Z score thresholds that include a range from the original image (0) to a high value (4), as well as Z scores that correspond to commonly used p-values (eg, 1.65= 0.1, 1.96=0.05, 2.58=0.01).
